### 工资项初始值

```python
# rt --> payroll result table
rt = { 
    'wt_1000': 10000, # CNY amount
    'wt_2001': 800,
    'wt_7070': -1500,
    'wt_801': None,
}
```

### 获取折减因子

假设全月排班日为23天   
月中入职后实际有效工作天数为16天   
折减因子/801计算逻辑如下:

```python
monthly_planned_working_days = 23
active_working_days = 16

print(r'Before factoring, /801 =', rt['wt_801'])
rt['wt_801'] = active_working_days / monthly_planned_working_days
print(r"After factoring, /801 =", rt['wt_801'])
```

对相关工资项进行折减计算如下：

```python
rt['wt_1000'] = round(rt['wt_1000'] * rt['wt_801'], 2)
```

折减步骤完成后当前RT表中工资项结果为:

```python
print(rt)
```
