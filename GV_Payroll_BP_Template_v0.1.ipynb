## General Tracking of Employees

### GlobalView Personal Number Range

According to 3M Global template document, once 3M employee is hired into ADP GlobalView system, a new SAP Personnel Number (called GlobalView PERNR) will be created in ADP as external numbering. The SAP personnel number is a numeric 8-digit field in the system. To keep different countries separated from each other, different number ranges will be set up. It will be automatically generated with the country modifier MOLGA code (2-digit numeric field) as the leading two digits of the personnel number / number range. This means that when an employee is first hired in China, a new personnel number is automatically created starting with ’28’ which is the MOLGA country code for China. Apart from the country identifiers, the personnel numbers are not descriptive.

| from number | to number |
|--|--|
|28000000|28999999|

### Cross-country Transfer

The GlobalView Payroll Number is country-specific, so that if an employee is transferred to a new country a new GV PERNR number will be generated.  In the event of return to a country in which they have previously been employed, the previous GV PERNR will be re-used.

- Employees in their home country
> Employees who exist in the system under the country they are originally from/ declare as their base country (Home Country) will be considered Expatriates (Expat)
- Employees in their host country
> Employees who exist in the system under the country they are currently working under (Host Country) will be considered Inpatriates (Inpat)
For 3M, the global ID number stored in infotype 0032.


### Domestic Transfer

When an employee is transferred to a new company code within China, the previous GV ID will be reused.

### Rates of Pay

Rates of Pay refers to any rate-based calculation within the system.  These are typically calculations such as overtime and leave calculations.

In this project, there are rates as following:

````python
ratesOfPayRulesDf = get_ratesOfPay_rule(bpcw_path)
ratesOfPayRulesDf
````

````python
vb001_ingredients = ratesOfPayRulesDf.loc[:,['Composed by Wage Types']][ratesOfPayRulesDf['Valuation Basis'] == r'/001']
vb001_ingredients = vb001_ingredients.iloc[0,0]
vb001_ingredients = vb001_ingredients.replace(' ', '')
vb001_ingredients = vb001_ingredients.split(',')
print(r'Ingredients of valuation basis /001 are:')
vb001_ingredients
````

Then formula of /001 will be:
  

> wt_/001 = $\sum$ {{vb001_ingredients}} / 174

According to above rule of rates of pay, we give out below example or /001:  

```
    Employee number = 28000000 
    wage type 1000 = 8000 CNY (1st sub-period)
    wage type 1000 = 9000 CNY (2nd sub-period)
    wage type 3000 = 3500 CNY (2nd sub-period)
    wage type 9001 = 150 CNY (1st sub-period)
    fixed divisor for valuation basis /001 is 174 hrs/month
```

````python
p28000000 = Person('28000000')

p28000000.it = [
    ItItem(1,'1000',8000),
    ItItem(2,'1000',9000),
    ItItem(2,'3000',3500),
    ItItem(1,'9001',150),
    ItItem(2,'4002',500)
]

print('Above codes creates sample person 28000000 and initializes "it" table as:')
it_df_28000000 = pd.DataFrame(p28000000.it)
it_df_28000000
````

Since now we have 2 sub-periods in the it table (separated by wpbp\_ind), GV will generate two wt_/001 records for each sub-period. For each sub-period, the formula:

> wt_/001 = $\sum$ {{vb001_ingredients}} / 174

will be executed.

Below codes simulates the calculation:

````python
filtered = it_df_28000000[it_df_28000000['wage_type'].isin(vb001_ingredients)]
print(r'Entries in "it" table that take part into /001 calculation are filtered by wage type:')
filtered
````

Grouped sum-up the amount by wpbp_ind column, and then add wt_/001 into "it" table:

````python
grouped = filtered['amount'].groupby(filtered['wpbp_ind'])
summed = grouped.sum()

print('Total amount of sub-period 1 & 2 are:')
print(summed)

for i in summed.index:
    item = ItItem(i,r'/001', rate=summed[i]/174)
    p28000000.it.append(item.__dict__)

it_df_28000000 = pd.DataFrame(p28000000.it)
it_df_28000000
````

When evaluating those rate-based wage types, the number in a certain sub-period will be multiplied with the wt_/001 rate in the same sub-period.
